**Perform below joins between cust_data & cust_demo.
a. Create data set â€œcust_leftjoinâ€ using Left Join
b. Create data set â€œcust_rightjoinâ€ using right Join
c. Create data set â€œcust_innterjoinâ€ using inner Join
d. Create data set â€œcust_fulljoinâ€ using full Join**

```python
Cust_lj=Cust_data.join(Cust_Demo,on='ID',how='left',lsuffix='cst_data')
print(len(Cust_lj))
```

```python
Cust_rj=Cust_data.join(Cust_Demo,on='ID',how='right',rsuffix='cst_Demo')
print(len(Cust_rj))
```

```python
Cust_ij=Cust_data.join(Cust_Demo,on='ID', how='inner',lsuffix='cst_Demo')
print(len(Cust_ij))
```

```python
Cust_oj=Cust_data.join(Cust_Demo,on='ID',how='outer',lsuffix='cst_Demo')
print(len(Cust_oj))
Cust_oj.columns
```

**Create summary report with below column by combination of Gender, Serious_delinquency.
a. Number of customers
b. Percentage of customers
c. Average Revolving utilization
d. Standard deviation with in the utilization
e. Average monthly income
f. Maximum Monthly income
g. Standard deviation with in the income
h. Average Age
i. Average Dependents**

```python
Cust_Grpd=Cust_oj.groupby(['Gender','SeriousDlqin2yrs']).agg(['count','mean','max','std'])[['IDcst_Demo','RevolvingUtilization','MonthlyIncome','age','No_of_dependents']]
Cust_summary=pd.DataFrame()
Cust_summary['Number_of_Customers']=Cust_Grpd['IDcst_Demo']['count']
total=Cust_summary['Number_of_Customers'].sum()
Cust_summary['Percentage_of_customers']=((Cust_summary['Number_of_Customers']/total)*100)
Cust_summary[['RevolvingUtilization_Average','RevolvingUtilization_Std']]=Cust_Grpd['RevolvingUtilization'][['mean','std']]
Cust_summary[['Monthly_Income_Average','Monthly_Income_max','Monthly_Income_std']]=Cust_Grpd['MonthlyIncome'][['mean','max','std']]
Cust_summary['Age_Average']=Cust_Grpd['age']['mean']
Cust_summary['Dependents_Average']=Cust_Grpd['No_of_dependents']['mean']
Cust_summary
```
