### 6. In customer table, create a new variable called â€œcustomer value segmentâ€ using customer value as follows. -  - High Value Segment - > 25000  - Medium Value Segment â€“ Between 10000 and 25000  - Low Value Segment â€“ less than or equal to 10000 

```python
conditions = [
    (cust['Customer_Value'] > 25000),
    (cust['Customer_Value'] <= 25000) & (cust['Customer_Value'] >= 10000),
    (cust['Customer_Value'] < 10000)]
choices = ['High Value Segment','Medium Value Segment','Low Value Segment']
cust['Customer_Value_Segment']= np.select(conditions, choices, np.nan)
cust
```

### 7. Create variables â€œaverage revenue per tripâ€ and â€œbalance pointsâ€ in the customer data set.  

```python
cust['Average_revenue_per_trip'], cust['Balance_points']=[(cust['Customer_Value']/cust['Buy_Times']), (cust['Points_Earned']-cust['Points_Redeemed'])]
cust
```

### 9. Calculate percentage of sales by each last city, state and region.

```python
#I have assumed the percentange for sales in total dataframe
total=cust['Customer_Value'].sum()
cust1=(cust.groupby(['Last_region','Last_state','Last_city']).sum()['Customer_Value']/total)*100
cust1
```

### 10. What is the count of customers, average number of purchases and average purchase transaction value by last state and city. 

```python
cust_2=cust.groupby(['Last_state','Last_city']).aggregate(['mean'])
cust_3=cust.groupby(['Last_state','Last_city']).aggregate(['count'])
cust_final=pd.DataFrame()
cust_final[['Customers_count']]=cust_3[['Customer_ID']]
cust_final[['Average_no_purchases','Average_Purchase_Transaction']]=cust_2[['Buy_Times','Customer_Value']]
cust_final
```
